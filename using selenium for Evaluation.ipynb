```python
#Q1: In this question you have to scrape data using the filters available on the webpage You have to use the 
#location and salary filter.
#You have to scrape data for ‚ÄúData Scientist‚Äù designation for first 10 job results.
#You have to scrape the job-title, job-location, company name, experience required.
#The location filter to be used is ‚ÄúDelhi/NCR‚Äù. The salary filter to be used is ‚Äú3-6‚Äù lakhs
# https://www.naukri.com/
```

```python
driver.get("https://www.naukri.com/")
```

```python
designation = driver.find_element(By.CLASS_NAME,"suggestor-input ")

designation.send_keys('Data Scientist')
```

```python
search = driver.find_element(By.CLASS_NAME,"qsbSubmit")

search.click()
```

```python
# after that go to webpage and select location delhi/ncr filter box and salary 3-6 lakhs.
```

```python
job_title = []
job_location = []
company_name = []
exper_require = []
```

```python
title_tags = driver.find_elements(By.XPATH,'//a[@class="title "]')

for i in title_tags:
    title=i.text
    job_title.append(title)
    job_title[:10]
    
    

location_tags = driver.find_elements(By.XPATH,'//span[@class="locWdth"]')

for i in location_tags:
    location=i.text
    job_location.append(location)
    job_location[:10]
    
    

company_tags = driver.find_elements(By.XPATH,'//div[@class=" row2"]/span/a[1]')

for i in company_tags:
    company=i.text
    company_name.append(company)
    company_name[:10]
    
    
    
experience_tags = driver.find_elements(By.XPATH,'//span[@class="expwdth"]')

for i in experience_tags:
    exp=i.text
    exper_require.append(exp)
    exper_require[:10]
```

```python
len(job_title[:10]),len(job_location[:10]),len(company_name[:10]),len(exper_require[:10])
```

```python
df = pd.DataFrame({'Title':job_title[:10],'Location':job_location[:10],'Company_Name':company_name[:10],'Experience':exper_require[:10]})

df
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://www.azquotes.com/")
```

```python
quotes = []
quotes_author = []
quote_types = []
```

```python
quote_tags = driver.find_elements(By.XPATH,'//a[@class="title"]')

for i in quote_tags:
    quote=i.text
    quotes.append(quote)
    
    
author_tags = driver.find_elements(By.XPATH,'//div[@class="author"]')

for i in author_tags:
    author=i.text
    quotes_author.append(author)
    
    
type_tags = driver.find_elements(By.XPATH,'//div[@class="tags"]')

for i in type_tags:
    types=i.text
    quote_types.append(types)
```

```python
len(quotes),len(quotes_author),len(quote_types)
```
